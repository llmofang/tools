```python
py=pd.read_csv('./拼音汉字映射表.csv')
```

```python
for i in range(len(py)):
    py.iloc[i]['py']=py.iloc[i]['py'][0]
```

```python
pygroup=py.groupby(['py'])
```

```python
data=[]
for name,group in pygroup:
    charas=''
    for i in range(len(group)):
        charas=charas+group.iloc[i]['characters']
    data.append([name,charas])
```

```python
data
```

```python
py1=pd.DataFrame(data,columns=['py','characters'])
```

```python
py1.to_csv('./汉字拼音简写对照.csv')
```

```python
data2=[]
for item in data:
    for it in item[1]:
        data2.append([it,item[0]])
```

```python
py2=pd.DataFrame(data2,columns=['characters','py'])
py2.to_csv('./汉字拼音简写对照.csv')
```

```python
dataArrAll=[]
for j in range(len(df)):
    pystrarr=['']
    for item in df.iloc[j]['name']:
        tempArr=[]
        rtdata=py2[py2['characters']==item]
        if rtdata is not None:
            for i in range(len(rtdata)):
                for pystr in pystrarr:
                    pystr=pystr+rtdata.iloc[i]['py']
                    tempArr.append(pystr)
        pystrarr=tempArr
    dataArr=[]
    dataArr.append(df.iloc[j]['code'])
    dataArr.append(df.iloc[j]['name'])
    dataArr.append(pystrarr) 
    dataArrAll.append(dataArr)
```
