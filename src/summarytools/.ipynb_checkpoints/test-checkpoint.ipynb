```python
sdf = pd.read_csv('./data.csv')
    sdf = sdf.rename(columns={sdf.columns[0]: "time", sdf.columns[1]: "entrust_no",
                              sdf.columns[2]: "code", sdf.columns[3]: "stock_name",
                              sdf.columns[4]: "price", sdf.columns[5]: "num",
                              sdf.columns[6]: "deal_price", sdf.columns[7]: "deal_num", 
                              sdf.columns[8]: "withdraw", sdf.columns[9]: "status"})

    sdf = sdf[sdf['deal_num'] != 0]
    sdf[sdf.columns[0]] = pd.to_datetime(sdf[sdf.columns[0]])
    sdf.sort(columns=sdf.columns[0], ascending=True)
```

```python
ddf = pd.DataFrame(columns=['stock_name', 'code', 'open_time', 'open_price', 'open_num',
                            'close_time', 'close_price','close_num', 'profit', 'amount', 'efficiency', 'hold_time'])
```

```python
for index, row in sdf.iterrows():
    print row.code
    unfinished = ddf[(ddf['code'] == row.code) & (ddf['close_price'] == 0) & (ddf['open_num']*row.num < 0) ]
    if len(unfinished) == 0:
        ddf.loc[len(ddf)] = [row.stock_name, row.code, row.time, row.deal_price, row.deal_num, row.time, 0, 0, 0, 0, 0, 0]
    else:
        for index, unfinished_row in unfinished.iterrows():
            if abs(row.deal_num) > 100:
                unfinished_row.close_time = row.time
                unfinished_row.close_price = row.deal_price
                if abs(unfinished_row.open_num) < abs(row.deal_num):                    
                    unfinished_row.close_num = - unfinished_row.open_num
                    row.deal_num = row.deal_num - unfinished_row.close_num
                else:
                    ddf.loc[len(ddf)] = [unfinished_row.stock_name, unfinished_row.code, unfinished_row.open_time, unfinished_row.open_price,
                                        unfinished_row.open_num + row.deal_num, unfinished_row.open_time, 0, 0, 0, 0, 0, 0]
                    unfinished_row.open_num = - row.deal_num
                    row.deal_num = 0
            else:
                break
        if abs(row.deal_num) > 100:
            ddf.loc[len(ddf)] = [row.stock_name, row.code, row.time, row.deal_price, row.deal_num, row.time, 0, 0, 0, 0, 0, 0]
```
